# CHAID ALGORITHM

```python
from CHAID import Tree, NominalColumn
import pandas as pd
import numpy as np

# Assuming you have obtained the selected features from logistic regression
selected_features = ['Internet_usg', 'school_work', 'job_performance', 'life_no_internet',
       'headache', 'appetite', 'sleep', 'fear', 'nervous', 'digestion',
       'troubled', 'unhappy', 'cry', 'enjoyment', 'decisions', 'work', 'play',
       'interest', 'suicide', 'tiredness', 'uncomfortable', 'easily_tired']

data_filtered = data1[selected_features + ['target']] 
target_variable_name = 'target'
independent_variable_columns = selected_features
dependent_variable_column = target_variable_name
tree = Tree.from_pandas_df(data_filtered, dict(zip(independent_variable_columns, ['nominal'] * len(independent_variable_columns))), dependent_variable_column,max_depth=3)

tree.print_tree()

cols = [NominalColumn(data_filtered[col].values, name=col) for col in independent_variable_columns]
dependent_column = NominalColumn(data_filtered[dependent_variable_column].values, name=dependent_variable_column)
tree = Tree(cols, dependent_column)

tree.print_tree()

lib_tree = tree.to_tree()
print(lib_tree)

# Access different nodes of the tree
#first_node = tree.tree_store[0]

# Access properties of the node
#print(first_node.members)
#print(first_node.split.p)
#print(first_node.split.score)
```
